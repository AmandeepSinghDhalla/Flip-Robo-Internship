## Question 2
Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.

```python
# Loading the pages and asking for request
# page2a gives top rated movies from 1 to 50
page2a = requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")
print(page2a)
# page2b gives top rated movies from 51 to 100
page2b = requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt")
print(page2b)

#Loading both pages into beautiful soup to get the page content
soup2a = BeautifulSoup(page2a.content)
soup2b = BeautifulSoup(page2b.content)

# Finding all the names of the movies in page2a and page2b
name2a = soup2a.find_all('h3', class_="lister-item-header")
name2b = soup2b.find_all('h3', class_="lister-item-header")

# Making a name list
Name = []

# appending the names from page2a in the list one by one
for i in name2a:

    for j in i.find_all("a"):

        Name.append(j.text.replace('\n',''))

# appending the names from page2b in the list one by one
for i in name2b:

    for j in i.find_all("a"):

        Name.append(j.text.replace('\n',''))

print(f"{Name}\n\nTotal Movies :{len(Name)}")
```

```python
# Finding movies ratings from page2a and page2b
rating2a = soup2a.find_all('div', class_="inline-block ratings-imdb-rating")
rating2b = soup2b.find_all('div', class_="inline-block ratings-imdb-rating")

# initialising movie Rating list
movie_rating = []

# Appending ratings from page2a in list
for i in rating2a:

    for j in i.find_all("strong"):

        movie_rating.append(j.text)

# Appending ratings from page2b in list
for i in rating2b:

    for j in i.find_all("strong"):

        movie_rating.append(j.text)

print(f"{movie_rating}\n\nTotal Ratings :{len(movie_rating)}")
```

```python
# Finding years of movie release from page2a and page2b
year2a = soup2a.find_all('span', class_="lister-item-year text-muted unbold")
year2b = soup2b.find_all('span', class_="lister-item-year text-muted unbold")

# initialising year list
Year = []

# appending the years in the list
for i in year2a:
    Year.append(i.text)
for i in year2b:
    Year.append(i.text)

print(f"{Year}\n\nTotal Years :{len(Year)}")
```

```python
Year[58] = "(2017)"
```

Now we save all the scraped data into the dataframe.

```python
# making the dataframe and saving all the lists into the dataframe. 
import pandas as pd
top_100_movies = pd.DataFrame()
top_100_movies['Movie Name'] = Name
top_100_movies['IMDB Rating'] = movie_rating
top_100_movies['Year Released'] = Year

print("The dataframe for top 100 movies of all time is: ")
top_100_movies
```

**Finally saving the dataframe into the csv.**

```python
top_100_movies.to_csv("Top 100 Movies All Time.csv")
```

 ## Question 3
 Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.

```python
page3 = requests.get("https://www.imdb.com/india/top-rated-indian-movies/?sort=ir,desc&mode=simple&page=1")
print(page3)

soup3 = BeautifulSoup(page3.content)

name3 = soup3.find_all('td', class_="titleColumn")

name_indian = []
for i in name3:
    for j in i.find_all('a'):
        name_indian.append(j.text)

rating3 = soup3.find_all('td', class_="ratingColumn imdbRating")
rating_indian = []
for i in rating3:
    for j in i.find_all('strong'):
        rating_indian.append(j.text)

year3 = soup3.find_all('span', class_="secondaryInfo")
year_indian=[]
for i in year3:
    year_indian.append(i.text)

top_100_indian_movies = pd.DataFrame()
top_100_indian_movies['Movie Name'] = name_indian
top_100_indian_movies['IMDB Rating'] = rating_indian
top_100_indian_movies['Year Released'] = year_indian
top_100_indian_movies
```

We need only top 100 Indian movies, so we removed rest of the movies.

```python
top_100_indian_movies = top_100_indian_movies.iloc[0:100,:]

# Saving to csv file.
top_100_indian_movies.to_csv("Top 100 Indian Movies")
```

## Question 4
Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’

```python
page4 = requests.get("https://bookpage.com/reviews?book_genre=fiction&page=1")
print(page4)

soup4 = BeautifulSoup(page4.content)

# Bookname scrapping.
name4 = soup4.find_all('h4', class_="italic")
bookname = []
for i in name4:
    for j in i.find_all('a'):
        bookname.append(j.text.replace('★','').replace(' ', ''))
#bookname

# Author name scrapping
author4 = soup4.find_all('p', class_="sans bold")
book_author = []
for i in author4:
    book_author.append(i.text.replace('\n', ''))
#book_author

# Book genre scrapping
genre4 = soup4.find_all('p', class_="genre-links hidden-phone")
book_genre = []
for i in genre4:
    book_genre.append(i.text.replace('\n', ''))
#book_genre

# Book Review scrapping
review4 = soup4.find_all('p', class_="excerpt")
book_review = []
for i in review4:
    book_review.append(i.text.replace('\n', ''))
#book_review
```

```python
books = pd.DataFrame()
books['Book Name'] = bookname
books['Author Name'] = book_author
books['Book Genre'] = book_genre
books['Book Review'] = book_review
books.sample(5)
```

Above we have the any 5 books with name, author, genre and review from bookpage.com

## Question 5
Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:

***i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.***

```python
page5a = requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
print(page5a)

soup5a = BeautifulSoup(page5a.content)

#Top Teams Scrapping
team5a = soup5a.find_all('span', class_="u-hide-phablet")
top_team5a = []
for i in team5a:
    top_team5a.append(i.text)

top_team5a = top_team5a[0:20]
# top_team5a

#Top team Matches and Points Scrapping
match_and_point5a = soup5a.find_all('td', class_="table-body__cell u-center-text")
matchpoint5a = []
for i in match_and_point5a:
    matchpoint5a.append(i.text)
# matchpoint5a

top_team_point5a = []
top_team_match5a = []
for i in range(len(matchpoint5a)):
    if i % 2 == 0:
        top_team_match5a.append(matchpoint5a[i])
    else:
        top_team_point5a.append(matchpoint5a[i])

banner_match5a = soup5a.find('td', class_="rankings-block__banner--matches")
banner_point5a = soup5a.find('td', class_="rankings-block__banner--points")
top_team_match5a.insert(0, banner_match5a.text)
top_team_point5a.insert(0, banner_point5a.text)

# Top teams Ratings scrapping
rating5a = soup5a.find_all('td', class_="table-body__cell u-text-right rating")
top_team_rating5a = []
for i in rating5a:
    top_team_rating5a.append(i.text)

banner_rating5a = soup5a.find('td', class_="rankings-block__banner--rating u-text-right")
banner_rating = []
banner_rating.append(banner_rating5a.text.replace('\n', ''))
banner_rating[0] = int(banner_rating[0])
banner_rating[0] = str(banner_rating[0])
#banner_rating

top_team_rating5a.insert(0, banner_rating[0])
top_team_rating5a

print(len(top_team5a), len(top_team_point5a), len(top_team_match5a), len(top_team_rating5a))
```

```python
top_10_ODI_teams = pd.DataFrame()
top_10_ODI_teams['Team Name'] = top_team5a
top_10_ODI_teams['Matches Played'] = top_team_match5a
top_10_ODI_teams['Total Points'] = top_team_point5a
top_10_ODI_teams['Team Rating'] = top_team_rating5a
top_10_ODI_teams
```

We only need data for the top 10 teams, so we remove rest of the data and save final dataframe into a csv file. **So top 10 ODI teams with matches, points and ratings are**

```python
top_10_ODI_teams = top_10_ODI_teams.iloc[0:10,:]
```

```python
top_10_ODI_teams.to_csv("Top 10 ODI Teams.csv")
```

## Question 8
Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
page8 = requests.get("https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YJfQb7UzZPY")
print(page8)

soup8 = BeautifulSoup(page8.content)

# Period Scraping
period8 = soup8.find_all('div', class_="col-sm-2 forecast-label")
period = []
for i in period8:
    period.append(i.text)
#period

# Short Description Scraping
period_desc8 = soup8.find_all('div', class_='col-sm-10 forecast-text')
period_desc = []
for i in period_desc8:
    period_desc.append(i.text)
#period_desc

# Temperature Scraping
temp8a = soup8.find_all('p', class_ = 'temp temp-high')
temp8b = soup8.find_all('p', class_ = 'temp temp-low')
temp = []
for i in temp8a:
    temp.append(i.text)
#temp
temp8 = []
for i in temp8b:
    temp8.append(i.text)
#temp8
temp_n = temp[0:4]
#temp_n
res_temp = []
for i in range(0, len(temp_n)):
    res_temp.append(temp_n[i])
    res_temp.append(temp8[i])
#res_temp
res_temp.append(temp[4])
from itertools import repeat
res_temp.extend(repeat('Not Available', 4)) # to add 4 more values such that we don't have NaN values.
#res_temp

# Temperature Description Scrapping
temp_desc8 = soup8.find_all('p', class_='short-desc')
temp_desc = []
for i in temp_desc8:
    temp_desc.append(i.text)
temp_desc.extend(repeat('Not Available', 4))
#temp_desc
```

```python
weather_forecast = pd.DataFrame()
weather_forecast['Upcoming Period'] = period
weather_forecast['Description'] = period_desc
weather_forecast['Temperature'] = res_temp
weather_forecast['Weather Description'] = temp_desc
weather_forecast
```

```python
weather_forecast.to_csv('San Francisco 7 Days Weather Forecast.csv')
```
