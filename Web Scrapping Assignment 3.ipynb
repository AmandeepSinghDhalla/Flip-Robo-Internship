# Q2
In the above question, now scrape the following details of each product listed in first 3 pages
of your search results and save it in a dataframe and csv. In case if any product vertical has
less than 3 pages in search results then scrape all the products available under that product
vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of
Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details"
and “Product URL”. In case, if any of the details are missing for any of the product then

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
# saving the 1st page url
page_url2a = 'https://www.amazon.in/s?k=guitar&ref=nb_sb_noss_1' 

# saving the products links in a list to open one by one for scrapping
url2a = []
product_url2a = driver1.find_elements_by_xpath("//a[@class = 'a-link-normal a-text-normal']")
for i in product_url2a:
    url2a.append(i.get_attribute("href"))

# making list for append all the details from 1st webpage of search results
product2a = []
rating2a = []
n_rating2a = []
price2a = []
delivery2a = []
return2a = []


for i in url2a:
    driver1.get(i)
    time.sleep(3)
    
# scraping product name from 1st page
    product_name2a = driver1.find_element_by_xpath("//span[@id = 'productTitle']")
    product2a.append(product_name2a.text)
    
# scraping total number of ratings from 1st webpage
    try:
        number_ratings2a = driver1.find_element_by_xpath("//span[@id = 'acrCustomerReviewText']")
        n_rating2a.append(number_ratings2a.text.replace(' ratings', ''))
    except NoSuchElementException: # if not present then append '-' character
        n_rating2a.append('-')
        
# scraping price of product from 1st page
    try:
        product_price2a = driver1.find_element_by_xpath("//span[@id = 'priceblock_ourprice']")
        price2a.append(product_price2a.text.replace('₹ ', ""))
    except NoSuchElementException:
        price2a.append('-')

# scraping average rating of product from 1st page    
    try:
        product_rating2a = driver1.find_element_by_xpath("//span[@data-hook = 'rating-out-of-text']")
        rating2a.append(product_rating2a.text.replace(' out of 5', ''))
    except NoSuchElementException:
        rating2a.append('-')

# scraping delivery details from 1st page
    try:
        product_delivery2a = driver1.find_element_by_xpath("//div[@id = 'ddmDeliveryMessage']")
        delivery2a.append(product_delivery2a.text.replace(' Details', ''))
    except NoSuchElementException:
        delivery2a.append('-')
    
# scraping return policy from 1st page
    try:
        product_return2a = driver1.find_element_by_xpath("//div[@id = 'RETURNS_POLICY']")
        return2a.append(product_return2a.text)
    except NoSuchElementException:
        return2a.append('-')
        
time.sleep(1)
driver1.get(page_url2a)
```

```python
# first finding the urls for each smartphone given in the webpage.
products4 = driver4.find_elements_by_xpath("//a[@class='_1fQZEK']")
urls4 = []
for i in products4:
    urls4.append(i.get_attribute("href"))
len(urls4)
```

As we see we have a total of 5 phone on the webpage, so we scrap all the details for each of these 5 smartphones.

```python
name4 = [] # making empty lists to append name of smartphone
price4 = [] # making empty lists to append price of smartphone
elements = [] # making empty lists to append other details of smartphone

for i in urls4: # loop to open each url in list 'urls4' one by one, then scrap all the details
    driver4.get(i)
    time.sleep(2)
    read_more_button = driver4.find_element_by_xpath("//button[@class = '_2KpZ6l _1FH0tX']")
    read_more_button.click()
    time.sleep(0.5)
    
    phone_name4 = driver4.find_element_by_xpath("//span[@class = 'B_NuCI']")
    phone_price4 = driver4.find_element_by_xpath("//div[@class = '_30jeq3 _16Jk6d']")
    element = driver4.find_elements_by_xpath("//li[@class = '_21lJbe']")
    
    
    name4.append(phone_name4.text) # appending names in the list
    price4.append(phone_price4.text) # appending price in the list
    for j in element:
        elements.append(j.text) # appending all the other elements in the list

driver4.close()

print(len(name4), len(price4), len(elements))
```

# Q5
Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on
google maps. For this problem, we will ask the user to search for the city and then give the coordinates.

```python
# Defining a function to find the coordinates for any city.
def find_coordinates(city : str):

    """Takes city as the input in string format that we want to search"""
    
    driver5 = webdriver.Chrome(driver_path)
    url5 = 'https://www.google.com/maps'
    driver5.get(url5)
    driver5.maximize_window()
    time.sleep(0.5)
    
    search_bar5 = driver5.find_element_by_xpath("//input[@aria-label = 'Search Google Maps']")
    search_bar5.send_keys(city, Keys.ENTER)
    time.sleep(2.5)

    city_search_url = driver5.current_url
    driver5.close()
    
    print(f"The map url for searched city is\n{city_search_url}\n")

    coordinates = city_search_url.split('/')
    coordinates = coordinates[6]
    coordinates = coordinates.replace('@', '').replace('z', '')
    coordinates = coordinates.split(',')
    latitude = coordinates[0]
    longitude = coordinates[1]
    print(f"Longitute of {city} is: {longitude}\nLatitude of {city} is: {latitude}")
```

Now we will use the above function to scrap the coordinates for any city in the world. It will take the city that we want to search as the input.

```python
find_coordinates('Durg')
```

```python
find_coordinates('Nagpur')
```

# Q6
Write a program to scrap details of all the funding deals for second quarter (i.e. July 20 –
September 20) from trak.in.

```python
driver6 = webdriver.Chrome(driver_path)
url6 = 'https://trak.in/'
driver6.get(url6)
driver6.maximize_window()
time.sleep(1)

open_funding_deals6 = driver6.find_element_by_id("menu-item-51510")
open_funding_deals6.click()
```

After inspecting the tables for July 2020 to September 2020, we see that each have unique id, so we use it to scrap the data.

```python
# Making empty lists to append the scraped data.
date6a = []
startupname6a = []
industryvertical6a = []
subvertical6a = []
location6a = []
investorname6a = []
investmenttype6a = []
amount6a = []


for i in range(48,51):
    # To scrap the date Date
    dates6 = driver6.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[2]'.format(i))
    for j in dates6:
        date6a.append(j.text)

    # To scrap the Startup Name
    startup_names6 = driver6.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[3]'.format(i))
    for j in startup_names6:
        startupname6a.append(j.text)
    
    # To scrap the Industry/Vertical
    industry_vertical6 = driver6.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[4]'.format(i))
    for j in industry_vertical6:
        industryvertical6a.append(j.text)
    
    # To scrap the Sub-Vertical
    sub_vertical6 = driver6.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[5]'.format(i))
    for j in sub_vertical6:
        subvertical6a.append(j.text)

    # To scrap the Location
    location6 = driver6.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[6]'.format(i))
    for j in location6:
        location6a.append(j.text)
    
    # To scrap the Investor
    investor_name6 = driver6.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[7]'.format(i))
    for j in investor_name6:
        investorname6a.append(j.text)
    
    # To scrap the Investment Type
    investment_type6 = driver6.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[8]'.format(i))
    for j in investment_type6:
        investmenttype6a.append(j.text)
    
    # To scrap the Amount
    amount6 = driver6.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[9]'.format(i))
    for j in amount6:
        amount6a.append(j.text)
        
driver6.close()
```

```python
funding_deals = pd.DataFrame()
funding_deals['Date'] = date6a
funding_deals['Startup Name'] = startupname6a
funding_deals['Industry Vertical'] = industryvertical6a
funding_deals['Sub-Vertical'] = subvertical6a
funding_deals['City/Location'] = location6a
funding_deals["Investor's Name"] = investorname6a
funding_deals['Investment Type'] = investmenttype6a
funding_deals['Amount(in $)'] = amount6a
funding_deals
```

```python
url7 = 'https://www.digit.in/'
driver7 = webdriver.Chrome(driver_path)
driver7.get(url7)
driver7.maximize_window()
time.sleep(1)

# to go to top 10 page
top_10_button = driver7.find_element_by_xpath("/html/body/div[1]/div[2]/div[4]/ul/li[4]/a")
top_10_button.click()
time.sleep(1)

# to go to laptop page
laptop_selection = driver7.find_element_by_xpath("/html/body/div[3]/div/div/div[2]/div[5]/div[1]/div/button[2]")
laptop_selection.click()
time.sleep(1)

# to go to gaming laptop page
gaming_laptop_selection = driver7.find_element_by_xpath('//*[@id="laptops"]/div[3]/a/div[2]')
gaming_laptop_selection.click()
```

```python
# Scraping for the names of the laptops
name7 = []
names = driver7.find_elements_by_xpath("//div[@data-eventaction = 'NonAMPScroll']")
for i in names:
    name7.append(i.text.replace('\n', ''))
#name7

# scraping the operating system of laptops
os7 = []
operatingsystem = driver7.find_elements_by_xpath("//div[@class = 'Spcs-details']//tr[3]//td[3]")
for i in operatingsystem:
    os7.append(i.text)
#os7

# scraping the display resolution of laptops
display7 = []
display = driver7.find_elements_by_xpath("//div[@class = 'Spcs-details']//tr[4]//td[3]")
for i in display:
    display7.append(i.text)
#display7

# scraping the processor of laptops 
processor7 = []
processor = driver7.find_elements_by_xpath("//div[@class = 'Spcs-details']//tr[5]//td[3]")
for i in processor:
    processor7.append(i.text)
#processor7

# scraping the memory of laptops
memory7 = []
memory = driver7.find_elements_by_xpath("//div[@class = 'Spcs-details']//tr[6]//td[3]")
for i in memory:
    memory7.append(i.text)
#memory7

# scraping the weight of laptops
weight7 = []
weight = driver7.find_elements_by_xpath("//div[@class = 'Spcs-details']//tr[7]//td[3]")
for i in weight:
    weight7.append(i.text)
#weight7

# scraping the dimension of laptops
dimension7 = []
dimension = driver7.find_elements_by_xpath("//div[@class = 'Spcs-details']//tr[8]//td[3]")
for i in dimension:
    dimension7.append(i.text)
#dimension7

# scraping the price of laptops
price7 = []
price = driver7.find_elements_by_xpath("//td[@class = 'smprice']")
for i in price:
    price7.append(i.text)
#price7

driver7.close()
```
